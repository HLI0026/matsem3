Projekt je zaměřen na vytvoření triangulace nad $[0,1]^2$, jako elementy se zde volí P2 prvky, pomocí `get_p2_triang`.

Konstrukce probíhá následně:
* vytvoření P1 prvků s vhodnou indexací, která umožní využít
    ```
    x = np.linspace(-1, 1, N)
    y = np.linspace(-1, 1, N)
    X, Y = np.meshgrid(x, y)
    V = np.column_stack([X.ravel(), Y.ravel()])
    ```
    dále tvoříme horní dolní a horní trojúhelník v rámci subčtverce zároveň pomocí:
    ```
    for i in range(k):
        for j in range(k):
            d0 = i * N + j
            d1 = d0 + 1
            d2 = d0 + (N + 1)

            h0 = d0
            h1 = d1 + N 
            h2 = d2 - 1 
    ```
* poté dochází k vynásobení indexů vrcholů dvěma a vytvořením midpointů jako součtu indexů vrcholů před vynásobením 
* zároveň s vytvořením indexů dochází k výpočtu polohy daného midpointu jako průměru polohy sousedních 2 vrcholů v rámci grafu

Projekt dále obsahuje funci `plot_p2_mesh`, která využije output `get_p2_triang` a vykreslí všechny trojúhelníky i s midpointy a indexy vrcholů v rámci sítě.

Poslední funkce je `print_idx_coords_per_triag`, která vypíše index vrcholu v síti a jeho koordinace, je možnost zadat argument, který specifikuje, který trojúhelník se má vybrat pro vypsání, defaultně vypíše všechny
